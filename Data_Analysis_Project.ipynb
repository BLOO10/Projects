**Disturbution of ADR by Hotel Type visulaization**

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="hotel", y="adr", data=td, showfliers=True)

plt.title("Distribution of Average Daily Rate (ADR) by Hotel Type", fontsize=14)
plt.xlabel("Hotel Type")
plt.ylabel("Average Daily Rate (ADR)")
plt.ylim(0, 500)  # optional: to cut extreme outliers for clearer view
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

#Visualization

**Creating an analysis based on families and non-families**


```python
# Create a new column "is_family"
td["is_family"] = td.apply(lambda row: 1 if (row["children"] > 0) else 0, axis=1)

# Check counts of families vs non-families
print(td["is_family"].value_counts())

# Preview some rows
print(td[["adults", "children", "is_family"]].head(10))
```

```python
counts =td["is_family"].value_counts()  # 0 = Non-family, 1 = Family
```

```python
plt.figure(figsize=(6,5))
plt.bar(["Non-Family", "Family"], counts, color=["skyblue", "salmon"])
plt.title("Number of Family vs Non-Family Bookings", fontsize=14)
plt.ylabel("Number of Bookings")
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

**Customer count based of the type of booking**

```python
customer_counts = td["customer_type"].value_counts()
```

```python
# 3. Plot bar chart
plt.figure(figsize=(8,6))
plt.bar(customer_counts.index, customer_counts.values, color=["skyblue", "salmon", "lightgreen", "orange"])
plt.title("Bookings by Customer Type", fontsize=14)
plt.xlabel("Customer Type")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)

# 4. Show plot
plt.show()
```

**Creating the 3D visualization of overall hotel booking**

```python



# Create 3D scatter plot
fig = px.scatter_3d(
    td,
    x="lead_time",
    y="adr",
    z="stays_in_week_nights",
    color="is_canceled",
    opacity=0.7
)


# Customize axis titles
fig.update_layout(
    scene=dict(
        xaxis_title="Days in Advance",
        yaxis_title="ADR",
        zaxis_title="Length of Stay"
    ),
    title="Hotel Bookings: 3D Scatter"
)

fig.show()
```

#Data Analysis

```python
# Count of canceled vs non-canceled bookings
cancel_counts = td['is_canceled'].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(
    cancel_counts,
    labels=["Not Canceled", "Canceled"],
    autopct='%1.1f%%',
    colors=["lightgreen", "salmon"],
    startangle=90,
    explode=(0.05, 0.05)
)
plt.title("Percentage of Canceled vs Non-Canceled Bookings")
```

```python
family_counts = td['is_family'].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(
    family_counts,
    labels=["Non-Family", "Family"],
    autopct='%1.1f%%',
    colors=["skyblue", "salmon"],
    startangle=90,
    explode=(0.05, 0.05)
)
plt.title("Percentage of Family vs Non-Family Bookings")
plt.show()
```
